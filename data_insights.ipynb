### Initial Column (or Row) Clean-Up

#### Removing Two Columns
1. `OFF STREET NAME` -> lots of missing values here, and the data is scattered.
2. `LOCATION` -> we already have longitude/latitude data, `LOCATION` is just duplicating it. 

```python
df_clean = df.drop(columns=['LOCATION', 'OFF STREET NAME'])

updated_columns = df_clean.columns.tolist() ## confirming removal -> displaying updated column list

## Below, is to show the percentage of missing values in the columns that I have chosen to remove.

removed_missing_values = {'LOCATION':df['LOCATION'].isna().sum() / len(df) * 100, 
    'OFF STREET NAME': df['OFF STREET NAME'].isna().sum() / len(df) * 100
                         }

print(updated_columns)
print(removed_missing_values)
```

#### Renaming/Transforming Columns

1. I am renaming `CRASH DATE` and `CRASH TIME` to `DATE` and `TIME` -> just easier to read.
2. I will also convert `DATE` and `TIME` to the `datetime` format.
3. I will also replace any NaN values with `unknown`, just to handle missing values properly.

```python
df_clean = df_clean.rename(columns={'CRASH DATE': 'DATE', 'CRASH TIME': 'TIME'})
df_clean['DATE'] = pd.to_datetime(df_clean['DATE'], format='%m/%d/%y')
df_clean['TIME'] = pd.to_datetime(df_clean['TIME'], format='%H:%M:%S')

updated_names = df_clean.columns.to_list() ## updated column names -> converting into a list
df_clean.head()
```

```python
categorical_columns = ['BOROUGH', 'ON STREET NAME', 'CROSS STREET NAME', 'CONTRIBUTING FACTOR VEHICLE 1', 
                       'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3',
                       'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5',
                       'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2', 'VEHICLE TYPE CODE 3',
                       'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5'
                      ]

df_clean[categorical_columns] = df_clean[categorical_columns].fillna("unknown")


numerical_columns = ['ZIP CODE', 'LATITUDE', 'LONGITUDE']
df_clean[numerical_columns] = df_clean[numerical_columns].fillna(df_clean[numerical_columns].median())
```

### Number of accidents that occurred at each borough.

```python
## Here, I am counting the number of accidents/borough, replacing `unknown` with `others` as the instructions tell me to do so.

accidents_by_borough = df_clean['BOROUGH'].replace("unknown", "others").value_counts()

print(accidents_by_borough)
```

```python
import matplotlib.pyplot as plt

# I made a bar chart to visualize this statistics and to compare the number of accidents btwn each borough. 

plt.figure(figsize=(10, 5))
accidents_by_borough.plot(kind='bar')
plt.title("Number of Accidents By Borough (January - August 2020)")

plt.xlabel("Borough")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.7)
```

### Mapping Distribution of Accidents Based on Longitude/Latitude

```python
import matplotlib.pyplot as plt

## This is just to filter data to only include longitude and latitude values that are VALID for NYC.
df_filtered = df_clean[(df_clean["LATITUDE"].between(40.4, 40.9)) & (df_clean["LONGITUDE"].between(-74.3, -73.7))]


## I am factorizing the boroughs to assign unique colors.

borough_labels, borough_index = pd.factorize(df_filtered["BOROUGH"])

## Now, I am going to create a scatter plot of accidents by geo location, colored based on borough.

plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_filtered["LONGITUDE"], df_filtered["LATITUDE"], 
                 c=borough_labels, alpha=0.5, cmap='tab20', s=10)

## I also picked the 'tab20' colormap in matplotlib.

## Below is a legend that maps boroughs to colors

unique_borough = df_filtered["BOROUGH"].unique() 
legend_label = [f"{i} - {borough}" for i, borough in enumerate(unique_borough)]

## I wrote an if-statement to ensure that it has valid data, before trying to plot.

if not df_filtered.empty:
    plt.legend(handles=scatter.legend_elements()[0], labels=legend_label, title="Boroughs")
                 
## Here, I am setting axis limits to zoom into the city. 
                 
plt.xlim(-74.3, -73.7)
plt.ylim(40.5, 40.9)

## Plot Formation

plt.title("Distribution of NYC Traffic Accidents (January - August 2020)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

### Covariance between each pair of the columns

### Which month did the most number of accidents occur in?

```python
## First, I need to extract the month from the `DATE` column.
import calendar

df_clean['MONTH'] = df_clean['DATE'].dt.month

## Next, I will count the number of accidents per month.

accidents_by_month = df_clean['MONTH'].value_counts().sort_index()

## By instructions, I will convert month numbers to associated abbreviations using the calendar module,
## and the `month_abbr` method. 

print("These are the accidents by months: ")
print(accidents_by_month.rename(index=lambda x: calendar.month_abbr[x]))
```

- The data suggests that NYC traffic accidents dropped quite a bit in April 2020. This is pretty clear from the significant drop in accidents compared to both the earlier and later months of the year.
